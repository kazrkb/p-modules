### Apply function in df

```python
min = no_df['Total Marks'].min()
max = no_df['Total Marks'].max()

no_df['Scaled Marks'] = no_df['Total Marks'].apply(lambda x : (x-min)/(max-min))
```

```python
no_df
```

```python
# custom built function

def grading(marks):
    if marks >= 260:
        return 'A+'
    else:
        return 'A-'

no_df['Grade'] = no_df['Total Marks'].apply(grading)
```

```python
no_df
```

```python
def curving_marks(no_df):
    a = no_df['Data Structure Marks'] + 2
    b = no_df['Python Marks'] * 1.1
    c = no_df['Algorithm Marks'] * 1.5
    return a+b+c

no_df['Curved Total Marks'] = no_df.apply(curving_marks,axis=1)
```
